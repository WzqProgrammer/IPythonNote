# 图像的读取、显示与保存
## 读取图像

```python
img1 = cv.imread('Dataset/test_1.jfif', 0)    # 灰度图
img2 = cv.imread('Dataset/test_1.jfif', 1)    # 彩色图
print(type(img1), ' ', img1.shape)
print(type(img2), ' ', img2.shape)
```

```python
img1
```

## 显示图像
### cv.imshow()
显示窗口自适应图片大小

```python
cv.imshow('image1', img1)
# cv.waitKey(0)    # 参数表示等待多少毫秒，0表示等待键盘输入，按任意键继续运行程序
k = cv.waitKey(0) & 0xFF  # 中文和英文输入都可以

if k == 27:    # 按 Esc 键退出
    cv.destroyAllWindows()
elif k == ord('s'):   # 按 s 键退出
    cv.destroyAllWindows()   # 删除窗口
    print('已删除')
```

### cv.nameWindow先创建窗口
先创建窗口，后显示图像  
* 默认cv.WINDOW_AUTOSIZE：自动生成窗口大小
* cv.WINDOW_NORMAL：允许你调整窗口大小

```python
cv.namedWindow('image')
cv.imshow('image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

### matplotlib显示图像


```python
plt.imshow(img2)
```

该方法颜色会出问题，原因是cv2.imread读出来的图像是GBR的顺序存储，而matplotlib是RGB的顺序，使用以下方法转换

```python
imageRGB = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
plt.show()
```

## 保存图像

```python
cv.imwrite('Dataset/test_11.jpg', img1)
cv.imwrite('Dataset/test_12.jpg', img2)
```

# 绘图
## 画直线

```python
# 生成一张黑色的图像
img = np.zeros((512, 512, 3), np.uint8)
# 画出一条粗细为5px的蓝色线段
cv.line(img, (0, 0), (511, 511), (255, 0, 0), 5)
# cv.imshow('image', img)
# cv.waitKey(0)
# cv.destroyAllWindows()
imageRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
plt.show()
```

## 画矩形
参数：左上角和右下角坐标

```python
img = np.zeros((512, 512, 3), np.uint8)
cv.rectangle(img, (100, 400), (400, 128), (0, 255, 0), 3)
imageRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
plt.show()
```

## 画圆

```python
img = np.zeros((512, 512, 3), np.uint8)
cv.circle(img, (250, 250), 120, (100, 100, 255), 4)
imageRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
plt.show()
```

## 画椭圆
cv.ellipse(img, center, axes, angle, startAngle, endAngle, color\[, thickness\[, linetype\[, shift\]]])
axes:长轴和短轴的长度  
angle：整个椭圆沿顺时针方向旋转的角度  
startAngle, endAngle：椭圆弧面沿顺时针方向起始的角度和结束角度

```python
img = np.zeros((512, 512, 3), np.uint8)
cv.ellipse(img, (250, 250), (100, 40), 0, 0, 360, (100, 100, 255), 4)
imageRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
plt.show()
```
